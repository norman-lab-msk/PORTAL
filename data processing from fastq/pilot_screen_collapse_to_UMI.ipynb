```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
directory = "/data1/normantm/angel/Pscreen_data/1st_screen_essentials/new"
```

```python
GFP2_1 = pd.read_csv(f"{directory}/rep2_GFP_L1_summary.csv")
GFP2_2 = pd.read_csv(f"{directory}/rep2_GFP_L2_summary.csv")
```

```python
GFP2 = pd.concat([GFP2_1, GFP2_2], ignore_index=True).groupby(["protospacer_mapped", "barcode_mapped", "UMI"]).sum()
```

```python
GFP2_UMI = GFP2.groupby(["protospacer_mapped", "barcode_mapped"]).size().to_frame("UMI_count")
```

```python
mCherry2_1 = pd.read_csv(f"{directory}/rep2_mCherry_L1_summary.csv")
mCherry2_2 = pd.read_csv(f"{directory}/rep2_mCherry_L2_summary.csv")
```

```python
mCherry2 = pd.concat([mCherry2_1, mCherry2_2], ignore_index=True).groupby(["protospacer_mapped", "barcode_mapped", "UMI"]).sum()
```

```python
mCherry2_UMI = mCherry2.groupby(["protospacer_mapped", "barcode_mapped"]).size().to_frame("UMI_count")
```

```python
UMI2 = GFP2_UMI.merge(mCherry2_UMI,
                      left_index=True,
                      right_index=True,
                      suffixes=("_GFP", "_mCherry"),
                      how="outer").fillna(0)
```
